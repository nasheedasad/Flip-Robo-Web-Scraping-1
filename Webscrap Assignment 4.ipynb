# Q9

```python
driver9=webdriver.Chrome(r'C:\Users\delkl\Downloads\chromedriver.exe')
```

```python
url9='https://www.imdb.com/list/ls095964455/'
driver9.get(url9)
```

```python
url99=driver9.find_elements_by_xpath("//h3[@class='lister-item-header']//a")
url99
```

```python
imdb=[]
for i in url99:
    imdb.append(i.get_attribute('href'))
imdb
```

```python
movie_name=[]
for i in imdb:
    driver9.get(i)
    try:
        movie_title=driver9.find_element_by_xpath("//h1[@class='TitleHeader__TitleText-sc-1wu6n3d-0 dxSWFG']")
        movie_name.append(movie_title.text)
    except NoSuchElementException:
        movie_name.append('-')
movie_name
```

```python
year_span=[]
for i in imdb:
    driver9.get(i)
    try:
        year=driver9.find_element_by_xpath("//a[@class='ipc-link ipc-link--baseAlt ipc-link--inherit-color TitleBlockMetaData__StyledTextLink-sc-12ein40-1 rgaOW']")
        year_span.append(year.text)
    except NoSuchElementException:
        year_span.append('-')
year_span
```

```python
genre=[]
for i in imdb:
    driver9.get(i)
    try:
        genre_tag=driver9.find_element_by_xpath("//div[@class='ipc-chip-list GenresAndPlot__GenresChipList-cum89p-4 gtBDBL']")
        genre.append(genre_tag.text.replace('\n',' '))
    except NoSuchElementException:
        genre.append('-')
genre
```

```python
runtime=[]
for i in imdb:
    driver9.get(i)
    try:
        run_time=driver9.find_element_by_xpath("//ul[@class='ipc-inline-list ipc-inline-list--show-dividers TitleBlockMetaData__MetaDataList-sc-12ein40-0 dxizHm baseAlt']//li[4]")
        runtime.append(run_time.text)
    except NoSuchElementException:
        runtime.append('-')
runtime
```

```python
ratings=[]
for i in imdb:
    driver9.get(i)
    try:
        rating=driver9.find_element_by_xpath("//div[@class='AggregateRatingButton__Rating-sc-1ll29m0-2 bmbYRW']")
        ratings.append(rating.text.replace('\n',''))
    except NoSuchElementException:
        ratings.append('-')
ratings
```

```python
votes=[]
for i in imdb:
    driver9.get(i)
    try:
        vote=driver9.find_element_by_xpath("//div[@class='AggregateRatingButton__TotalRatingAmount-sc-1ll29m0-3 jkCVKJ']")
        votes.append(vote.text)
    except NoSuchElementException:
        votes.append('-')
votes
```

```python
print(len(movie_name),len(year_span),len(runtime),len(genre),len(ratings),len(votes))
```

```python
TV=pd.DataFrame({})
TV['Name']=movie_name
TV['Year Span']=year_span
TV['Runtime']=runtime
TV['Genre']=genre
TV['Ratings']=ratings
TV['Votes']=votes
```

```python
TV
```
