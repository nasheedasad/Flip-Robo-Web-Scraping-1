# Q10

```python
driver10=webdriver.Chrome(r'C:\Users\delkl\Downloads\chromedriver.exe')
```

```python
url10='https://www.hostelworld.com/'
driver10.get(url10)
```

```python
search_location=driver10.find_element_by_id('search-input-field')
search_location.send_keys('London')
```

```python
search_btnlocation=driver10.find_element_by_xpath("//button[@class='button primary large']")
search_btnlocation.click()
```

```python
url100=driver10.find_elements_by_xpath("//h2[@class='title title-6']//a")
url100
```

```python
hostels=[]
for i in url100:
    hostels.append(i.get_attribute('href'))
hostels
```

```python
name=[]
for i in hostels:
    driver10.get(i)
    try:
        name_title=driver10.find_element_by_xpath("//div[@class='title-2']")
        name.append(name_title.text)
    except:
        name.append('-')
name
```

```python
distance_tags=driver10.find_elements_by_xpath("//span[@class='description']")
distance=[]
for i in distance_tags:
    distance.append(i.text.replace('Hostel','').replace('Bed and Breakfast','').replace('-','').strip())
distance
```

```python
ratings=[]
for i in hostels:
    driver10.get(i)
    try:
        rating_title=driver10.find_element_by_xpath("//div[@class='score orange big']")
        ratings.append(rating_title.text)
    except:
        ratings.append('-')
ratings
```

```python
reviews=[]
for i in hostels:
    driver10.get(i)
    try:
        review_title=driver10.find_element_by_xpath("//div[@class='reviews']")
        reviews.append(review_title.text)
    except:
        reviews.append('-')
reviews
```

```python
overallreviews=[]
for i in hostels:
    driver10.get(i)
    try:
        overallreview_title=driver10.find_element_by_xpath("//ul[@class='rating-breakdown rating-grid']")
        overallreviews.append(overallreview_title.text.replace('\n',' '))
    except:
        overallreviews.append('-')
overallreviews
```

```python
privatedorm_tags=driver10.find_elements_by_xpath("//div[@class='price-col']")
privatedorm=[]
for i in privatedorm_tags:
    privatedorm.append(i.text.replace('\n',' '))
privatedorm
```

```python
private=[]
for i in range(0,len(privatedorm),2):
    private.append(privatedorm[i])
private
```

```python
dorm=[]
for i in range(1,len(privatedorm),2):
    dorm.append(privatedorm[i])
dorm
```

```python
facilities=[]
for i in hostels:
    driver10.get(i)
    try:
        facility_tags=driver10.find_element_by_xpath("//div[@class='facilities-container']")
        facilities.append(facility_tags.text.replace('\n',' '))
    except:
        facilities.append('-')
facilities
```

```python
description=[]
for i in hostels:
    driver10.get(i)
    try:
        desc_tags=driver10.find_element_by_xpath("//div[@class='description-container']")
        description.append(desc_tags.text.replace('\n',' '))
    except:
        description.append('-')
description
```

```python
print(len(hostels),len(distance),len(name),len(ratings),len(reviews),len(overallreviews),len(facilities),len(description),len(private),len(dorm))
```

```python
hostel=pd.DataFrame({})
hostel['Name']=name
hostel['Distance']=distance
hostel['Rating']=ratings
hostel['Total Reviews']=reviews
hostel['Overall Reviews']=overallreviews
hostel['Privates Starting From']=private
hostel['Dorms Starting From']=dorm
hostel['Facilities']=facilities
hostel['Property Description']=description
```
