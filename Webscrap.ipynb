# Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
from bs4 import BeautifulSoup
import requests

page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup=BeautifulSoup(page.content)
header=soup.find_all(['h1', 'h2','h3','h4','h5','h6'])
for i in header:
    print(i)
```

# Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).

```python
from bs4 import BeautifulSoup
import requests
```

```python
imdb=requests.get('https://www.imdb.com/list/ls091520106/')
soup=BeautifulSoup(imdb.content)
Name=soup.find_all('h3',class_='lister-item-header')
movie_name=[]
for i in Name:
    for j in i.find_all('a'):
        movie_name.append(j.text.replace('\n',''))
movie_name
```

```python
Rating=soup.find_all('div',class_='ipl-rating-star small')
movie_rating=[]
for i in Rating:
    movie_rating.append(i.text.replace('\n',''))
movie_rating
```

```python
len(movie_name)
```

```python
len(movie_rating)
```

```python
YOR=soup.find_all('span',class_='lister-item-year text-muted unbold')
Year_of_release=[]
for i in YOR:
    Year_of_release.append(i.text)
Year_of_release
```

```python
len(Year_of_release)
```

```python
import pandas as pd
movies=pd.DataFrame({})
movies['Name']=movie_name
movies['Rating']=movie_rating
movies['YOR']=Year_of_release
```

```python
movies
```

# Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release).

```python
from bs4 import BeautifulSoup
import requests

imdb1=requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
soup=BeautifulSoup(imdb1.content)
Name1=soup.find_all('td',class_='titleColumn')
movies_name=[]
for i in Name1:
    for j in i.find_all('a'):
        movies_name.append(j.text.replace('\n',''))
movies_name
```

```python
Rating1=soup.find_all('td',class_='ratingColumn imdbRating')
movies_rating=[]
for i in Rating1:
    movies_rating.append(i.text.replace('\n',''))
movies_rating
```

```python
Y_O_R=soup.find_all('span',class_='secondaryInfo')
Year_of_release1=[]
for i in Y_O_R:
    Year_of_release1.append(i.text)
Year_of_release1
```

```python
import pandas as pd
movies1=pd.DataFrame({})
movies1['Name1']=movies_name
movies1['Rating1']=movies_rating
movies1['Y_O_R']=Year_of_release1
```

# Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’


```python
from bs4 import BeautifulSoup
import requests

book=requests.get('https://bookpage.com/reviews')
soup=BeautifulSoup(book.content)
Book=soup.find_all('h4',class_='italic')
book_titles=[]
for i in Book:
    book_titles.append(i.text.replace('\n',''))
book_titles
```

```python
Author=soup.find_all('p',class_='sans bold')
author_names=[]
for i in Author:
    author_names.append(i.text.replace('\n',''))
author_names
```

```python
Genre=soup.find_all('p',class_='genre-links hidden-phone')
genre_names=[]
for i in Genre:
    genre_names.append(i.text.replace('\n',''))
genre_names
```

```python
Review=soup.find_all('p',class_='excerpt')
review_names=[]
for i in Review:
    review_names.append(i.text.replace('\n',''))
review_names
```

```python
import pandas as pd
books=pd.DataFrame({})
books['Book']=book_titles
books['Author']=author_names
books['Genre']=genre_names
books['Review']=review_names
```

```python
books
```

# Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:

# i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
from bs4 import BeautifulSoup
import requests

icc=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup=BeautifulSoup(icc.content)
Team=soup.find_all('td',class_='table-body__cell rankings-table__team')
team_titles=[]
for i in Team:
    for j in i.find_all('span',class_='u-hide-phablet'):
        team_titles.append(j.text.replace('\n',''))
team_titles
```

# Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
from bs4 import BeautifulSoup
import requests

page1=requests.get('https://internshala.com/fresher-jobs')
soup=BeautifulSoup(page1.content)
Titles=soup.find_all('div',class_='heading_4_5 profile')
job_titles=[]
for i in Titles:
    job_titles.append(i.text.replace('\n',''))
job_titles
```

```python
Company=soup.find_all('a',class_='link_display_like_text')
companies=[]
for i in Company:
    companies.append(i.text.replace('\n','').strip())
companies
```

```python
len(job_titles)
```

```python
len(companies)
```

```python
CTC=soup.find_all('div',class_='internship_other_details_container')
salary=[]
for i in CTC:
    salary.append(i.text.replace('\n','').strip())
salary
```

```python
len(salary)
```

```python
import pandas as pd
jobs=pd.DataFrame({})
jobs['Titles']=job_titles
jobs['Company']=companies
jobs['CTC']=salary
```

```python
jobs
```

# Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
from bs4 import BeautifulSoup
import requests

weather=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YQETMr0zaM9')
soup=BeautifulSoup(weather.content)
Period=soup.find_all('p',class_='period-name')
period=[]
for i in Period:
    period.append(i.text.replace('\n',''))
period
```

```python
Short_Description=soup.find_all('p',class_='short-desc')
description=[]
for i in Short_Description:
    description.append(i.text.replace('\n',''))
description
```

```python
import pandas as pd
weather=pd.DataFrame({})
weather['Period']=period
weather['Short_Description']=description
```
